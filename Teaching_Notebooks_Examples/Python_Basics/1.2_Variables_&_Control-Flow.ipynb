### Variables

Variables allow us to store data to use later in our program:

In python, creating a variable is simple, unlike in other languages with strict typing, we don't need to declare the variable's type when we create them:

```python
x = 2
x
```

You can create several variables at a time:

```python
a, b, c = [1,2,4], "Br", 22
a, b, c
```

We can recall variables created earlier:

```python
print(x, a, b, c)
```

We can change the content and type of a variable at will:

```python
y = 11
print(y)
y = "Cl"
print(y)
```

```python
my_list = [0,1,2,3,4]
my_list[0] = "first_element"
print(my_list)
```

### For Loops:

For loops allow us to loop over an iterator (e.g. a list):

```python
my_list = [0,1,2,3,4,5,6]

for number in my_list:
    print(number)
```

The `range()` function returns an iterator of the numbers between the specified ends: 

```python
for i in range(100, 110):
    print(i)
```

We can perform other operations during a for loop: 

```python
for i in range(10):
    print(i, i**2, (i**2)+i)
```

### While Loops:

While loops run until their condition is no longer true.

In this example, the loop will run while x is less than 10 and then will not run the next loop once it does reach 10. 

```python
x = 0
while x < 10:
    print(x)
    x += 1
```

The condition you evaluate doesn't have to be numeric! 

```python
my_string = "Chemistry is fun!"
while my_string != "":
    print(my_string)
    my_string = my_string[:-1]
```

In this loop we aren't incrementing our variable x, so x will never be greater than 100 and the loop will run forever! (Or until you stop in manually!) 

Try to avoid creating these situations accidentally. Though there are some times when you will want some code to keep running forever! 

```python
x = 0
while x > 100:
    print("Loop!")
```
