```python
import sys
sys.path.append("/home/jch936/Software/bmpga/")
from bmpga.potentials.Gupta_Implementation.guptaPotential import GuptaPotential
import numpy as np

import matplotlib.pyplot as plt 
```

```python
pot = GuptaPotential(["Zn", "Zn", "Zn"])
pot.calculate_total_energy(np.array([[-0.0025257515, -1.1132616080, -0.1490790579],
                                     [0.6823000831, -0.6720141616, 0.1308423179],
                                     [0.2779619653, -1.2199766042, 0.6583738879]]))
```

```python
plot_pot = GuptaPotential(["Zn", "Zn"]) 

rs = np.linspace(0.5, 2.0, 1000)

Es = []

for r in rs:
    Es.append(plot_pot.calculate_total_energy(np.array([[0,0,0], [0,0,r]])))
    
#print(rs, Es)
plt.plot(rs, Es)
plt.plot([min(rs), max(rs)], [0,0], "k:")

plt.xlabel("r$_{ij}$")
plt.ylabel("E / eV")

plt.show()
```
